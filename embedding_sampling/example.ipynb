```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, Normalizer

from utils import fast_dim_reduction, birch_clustering, fast_cluster_metric, optuna_birch_tuning, stratified_cluster_sampling, plot_optimization_history

# Generate synthetic data (already reduced to 9 dimensions)
np.random.seed(42)
num_samples = 25000
num_features = 256
fake_embeddings = np.random.randn(num_samples, num_features)
fake_embeddings = Normalizer().fit_transform(fake_embeddings)
fake_texts = [f"Document {i}" for i in range(num_samples)]

# Step 1: Use Optuna to tune Birch parametefrs
best_params, study = optuna_birch_tuning(
    fast_dim_reduction(
        StandardScaler().fit_transform(fake_embeddings), 
        method='pca', 
        n_components=12
    ), 
    n_trials=30, 
    metric='davies_bouldin',
    show_progress=True
)

# Plot optimization results
fig = plot_optimization_history(study)
#plt.show()
```

```python
# Step 2: Apply clustering with best parameters
labels = birch_clustering(
    fast_dim_reduction(
        StandardScaler().fit_transform(fake_embeddings), 
        method='pca', 
        n_components=12
    ), 
    **best_params
)

# Step 3: Sample documents using stratified cluster sampling
sample_size = max(250, len(np.unique(labels)) * 2)
sampled_indices = stratified_cluster_sampling(fake_texts, labels, sample_size, min_per_cluster=1)
sampled_texts = [fake_texts[i] for i in sampled_indices]

# Step 4: Display results
print(f"\nTotal documents: {len(fake_texts)}")
print(f"Number of clusters: {len(np.unique(labels))}")
print(f"Cluster distribution: {np.bincount(labels)}")

print(f"\nSampled {len(sampled_texts)} documents:")
for i, text in enumerate(sampled_texts):
    cluster_id = labels[sampled_indices[i]]
    print(f"  {text} (Cluster {cluster_id})")

# Analyze sampling quality
sampled_cluster_distribution = np.bincount([labels[i] for i in sampled_indices])
print("\nSampling statistics:")
print(f"  Original cluster distribution: {np.bincount(labels)}")
print(f"  Sampled cluster distribution: {sampled_cluster_distribution}")

# Calculate the percentage of each cluster that was sampled
for cluster_id in range(len(np.unique(labels))):
    original_count = np.sum(labels == cluster_id)
    sampled_count = np.sum([labels[i] == cluster_id for i in sampled_indices])
    percentage = (sampled_count / original_count) * 100 if original_count > 0 else 0
    print(f"  Cluster {cluster_id}: {sampled_count}/{original_count} documents ({percentage:.1f}%)")
```
